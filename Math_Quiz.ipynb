```python
import matplotlib.pyplot as plt
import random
import fractions

def display_fraction_question(numerator1, denominator1, numerator2, denominator2, operation):
    """Displays an adding or subtracting fraction question and returns the correct answer and choices."""
    if operation == '+':
        correct_numerator = numerator1 * denominator2 + numerator2 * denominator1
    else:  # operation == '-'
        correct_numerator = numerator1 * denominator2 - numerator2 * denominator1

    correct_denominator = denominator1 * denominator2
    correct_answer = fractions.Fraction(correct_numerator, correct_denominator).limit_denominator()

    # Generate answer choices, ensuring uniqueness
    answers = {correct_answer}
    while len(answers) < 4:
        random_numerator = random.randint(1, 50)
        random_denominator = random.randint(1, 50)
        answers.add(fractions.Fraction(random_numerator, random_denominator).limit_denominator())

    answers = list(answers)
    random.shuffle(answers)

    # Create a plot to display the fraction problem
    plt.figure(figsize=(5, 5))
    problem_text = rf'$\frac{{{numerator1}}}{{{denominator1}}}$ {operation} $\frac{{{numerator2}}}{{{denominator2}}}$'
    plt.text(0.5, 0.7, problem_text, fontsize=30, ha='center', va='center')
    plt.text(0.5, 0.4, 'What is the simplified fraction result?', fontsize=20, ha='center', va='center')

    # Display answer choices
    for i, answer in enumerate(answers):
        plt.text(0.5, 0.2 - i * 0.1, f'{i + 1}. {answer}', fontsize=18, ha='center', va='center')

    plt.axis('off')  # Hide axes
    plt.show()

    return correct_answer, answers

def fraction_quiz():
    """Runs a challenging fraction addition and subtraction quiz."""
    score = 0
    total_questions = 20
  # Increased difficulty

    print("Welcome to Kairo's Math Quiz for Grade 8\n")

    for question_num in range(1, total_questions + 1):
        print(f"Question {question_num}:")
        numerator1 = random.randint(1, 20)
        denominator1 = random.randint(1, 20)
        numerator2 = random.randint(1, 20)
        denominator2 = random.randint(1, 20)
        operation = random.choice(['+', '-'])

        correct_answer, answers = display_fraction_question(numerator1, denominator1, numerator2, denominator2, operation)

        try:
            user_choice = int(input("Enter the number of your answer choice: "))
            if 1 <= user_choice <= 4 and answers[user_choice - 1] == correct_answer:
                print("✅ Correct! Well done.\n")
                score += 1
            else:
                print(f"❌ Incorrect. The correct answer was {correct_answer}.")
        except ValueError:
            print("❌ Invalid input. Please enter a number between 1 and 4.\n")

    print(f"\nQuiz Over! Your final score: {score}/{total_questions}")
    if score == total_questions:
        print("🎉 Amazing job! You mastered it!")
    elif score >= total_questions // 2:
        print("😊 Good effort! Keep practicing.")
    else:
        print("📚 Don't give up! Keep working on fractions.")

# Run the quiz
if __name__ == "__main__":
    fraction_quiz()
```
